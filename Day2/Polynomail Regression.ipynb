<h1>Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

<h1>Loading Dataset

```python
df=pd.read_csv(r'C:\Users\ASUS\Desktop\Dataset1.csv')
```

```python
df.head()
```

```python
X=df.iloc[:,0].values
Y=df.iloc[:,1].values
```

```python
X=X.reshape(-1,1)
Y=Y.reshape(-1,1)
```

<h1>Visualising Data

```python
plt.scatter(X,Y)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

<h1>Spliting Data into train and test set

```python
from sklearn.model_selection import train_test_split
```

```python
train_x,test_x,train_y,test_y=train_test_split(X,Y,train_size=0.7,random_state=100)
```

<h1>Implementing Linear Regression

```python
from sklearn import linear_model
regr=linear_model.LinearRegression()
```

```python
regr.fit(train_x,train_y)
print(regr.coef_)
print(regr.intercept_)
```

```python
plt.scatter(X,Y)
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.plot(X,X*regr.coef_+regr.intercept_,'-r')
plt.show()
```

<h1>Polynomail Regression 
    <h3>Degree-3

```python
from sklearn.preprocessing import PolynomialFeatures

poly3 = PolynomialFeatures(degree=3)
train_x_poly3 = poly3.fit_transform(train_x)
train_x_poly3
```

```python
lin_poly3 = linear_model.LinearRegression()
lin_poly3.fit(train_x_poly3,train_y)
print ('Coefficients: ', lin_poly3.coef_)
print ('Intercept: ',lin_poly3.intercept_)
```

```python
plt.scatter(X,Y,  color='blue')
plt.plot(X, lin_poly3.predict(poly3.fit_transform(X)), color="red" )
plt.xlabel("x-axis")
plt.ylabel("y-axis")
```
