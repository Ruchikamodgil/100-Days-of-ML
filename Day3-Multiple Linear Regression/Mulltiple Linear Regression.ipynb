<h1>Importing Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

<h1>Loading Dataset

```python
df=pd.read_csv(r"C:\Users\ASUS\Desktop\50_Startups.csv")
```

```python
df.head()
```

```python
df.describe()
```

```python
X=df.iloc[:,:-1]
Y=df.iloc[:,-1]
```

<h1>Data Preprocessing

```python
snum=pd.get_dummies(X['State'],drop_first=True)
```

```python

snum
```

```python
X=X.drop('State',axis=1)
```

```python
X=pd.concat([X,snum],axis=1)
```

```python
X
```

<h1>Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
train_x,test_x,train_y,test_y=train_test_split(X,Y,train_size=0.7,random_state=100)
```

<h1>Implementing Multiple Linear Regression

```python
from sklearn import linear_model
regr=linear_model.LinearRegression()
```

```python
regr.fit(train_x,train_y)
print(regr.coef_)
print(regr.intercept_)
```

<h1>Predicting Output on Test Set

```python
pred_y=regr.predict(test_x)
```

```python
pred_y
```

<h1>Comparing Results

```python
results=pd.DataFrame({'Actual': test_y, 'Predicted': pred_y})
results
```

```python
from sklearn.metrics import r2_score
```

```python
print(r2_score(test_y,pred_y))
```

```python
print("Mean absolute error : %.2f" % np.mean(np.absolute(pred_y - test_y)))
```
